```python
txt = 'KoNLPy는 한국어 정보처리를 위한 파이썬 패키지입니다. 설치법은 이 곳을 참고해주세요. NLP를 처음 시작하시는 분들은 시작하기 에서 가볍게 기본 지식을 습득할 수 있습니다.'
```

```python
import nltk

# 정규식표현으로 토큰만들기 
words = nltk.regexp_tokenize(txt, r'\w+')
print(words[:5])
```

```python
# 문장으로 토큰 만들기
sents = nltk.sent_tokenize(txt)
sents
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tag import pos_tag

stemmer = PorterStemmer()
print([stemmer.stem(word) for word in words])
```

```python
# 변화가 없네? 
lemm = WordNetLemmatizer()
print([lemm.lemmatize(word, pos = 'v') for word in words])
```

```python
# 포스태그 해보면? 품사가 다 명사로 나옴 
pos_tag(words)
```

## 형태소 분석기 공통 메소드
- `morphs(string)` : 형태소 단위로 토큰화(tokenize)
- `nouns(string)` : 명사만 추출하여 토큰화(tokenize)    
- `pos(string)`: 품사 부착
    - 형태소 분석기 마다 사용하는 품사태그가 다르다.
        - https://konlpy-ko.readthedocs.io/ko/v0.5.2/morph/
- `tagset`: 형태소 분석기가 사용하는 품사태그 설명하는 속성. 

```python
from konlpy.tag import Okt, Komoran # Okt : Open Korea TExt 
```

```python
txt = '''[강병수 / 낚시꾼 구조 시민 : 거기 세 분 중에 다 내가 들어갈게. 내가 들어갈게. 이렇게 말씀하셨는데, 그래도 거기서 제가 나이가 제일 어려서, 제가 들어가겠다고 하고.]
무사히 구조를 마친 강 씨, 그런데 다시 바다로 뛰어듭니다.
뒤집힌 보트까지 꺼내옵니다.
구조를 지켜보던 일행이 화면으로 보기에도 이상하리 만치 침착했던 건 모두 수상안전 전문가였기 때문이었습니다.
구조된 남성은 일행이 가져온 옷으로 체온을 유지해 안정을 취했고, 무사히 귀가한 것으로 전해졌습니다.'''
```

```python
okt = Okt()

tokens = okt.morphs(txt) #  형태소 추출
len(tokens)

print(tokens)
```

```python
# 명사만 추출 
noun_tokens = okt.nouns(txt)
noun_tokens
len(noun_tokens)
```

```python
# 품사 부착 
pos_tokens = okt.pos(txt)
pos_tokens
```

```python
pos_tokens2 = okt.pos(txt, join= True)
print(pos_tokens2)
```

```python
okt.tagset # 품사 태그가 무엇인지 알수있다. 
```

```python
sample = '이것도 되나욬ㅋㅋㅋㅋㅋ'
okt.morphs(sample)
```

```python
okt.morphs(sample, norm = True) # norm은 트위터에만있는 기능 
```

```python
okt.morphs('반갑습니당', norm=True)
```

# KoNLPy 형태소 분석기와 NLTK를 이용한 문서의 분석

```python
from nltk import Text
from konlpy.corpus import kolaw
from konlpy.tag import Komoran
```

```python
law = kolaw.open('constitution.txt').read()
print(law[:100])
```

```python
# 명사만 추출 
kom = Komoran()
nouns = kom.nouns(law)
nouns
len(nouns)
```

```python
# 1글자인 단어들은 제거해보자 
nouns = [word for word in nouns if len(word)>1]
len(nouns)
```

```python
text = Text(nouns, name = '헌법')
text
```

```python
'대한민국 빈도수 : {}, 헌법 빈도수 : {}, 국민 빈도수 : {}'.format(text.count('대한민국'), text.count('헌법'), text.count('국민'))
```

```python
import matplotlib.pyplot as plt 
import matplotlib as mpl 
mpl.rcParams['font.family'] = 'AppleGothic'
```

```python
plt.figure(figsize = (10,10))

text.plot(20)
# plt.title('헌법에 나온 명사 빈도수 상위 20개')

plt.show()
```

```python
plt.figure(figsize = (10,10))

text.dispersion_plot(['법률','대통령','국가','국회','헌법'])
```

```python
# 매개변수로 전달한 단어가 나오는 라인을 출력 
# 국민이나오는 50글자 만큼 라인을 10줄을 출력한다. 
text.concordance('국민', width = 50, lines = 10)
```

```python
# 빈도수
freq = text.vocab()
freq
```

```python
f'총 단어수 : {freq.N()}, 고유단어 수 : {freq.B()}, 최빈단어 : {freq.max()}, 최빈단어의빈도수 : {freq.get(freq.max())}, 최반단어의 출연비율 : {freq.freq(freq.max())}'
```

```python
word_cnt = freq.most_common()
word_cnt
```

```python
import pandas as pd 
df = pd.DataFrame(word_cnt, columns = ['단어', '빈도수'])
df.head(10)
```

```python
df.tail(10)
```

```python
# 5번 이상 나온 단어들만 조회 
df2 = df[df['빈도수'] >= 5]
df2
```
