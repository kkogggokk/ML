## 62.3 코엔엘파이가제공하는 형태소분석기
- 형태소 분석기 사전이란?
    - 형태소 분석 함수들을 제공하는 모듈을 형태소 


- 코엔엘파이가 제공하는 5개의 형태소 분석기 클래스가 있다. 
    - 1) Hannanum(한나눔)
    - 2) Kkma(꼬꼬마)
    - 3) Komoran(코모란)
    - 4) Mecab(메카브)
    - 5) okt(Open Korean Text)
    - 비교해보기 : https://konlpy-ko.readthedocs.io/ko/v0.5.2/morph/


- 메소드는 아래와 같다. 
|번호|메소드|설명|
|--|--|--|
|1|morphs(str)|형태소 단위로 토큰화(tokenize)|
|2|nouns(str)|명사만 추출하여 토큰화(tokenize)|
|3|pos(str)|품사 부착|
|4|tagset|형태소 분석기가 사용하는 품사태그 설명하는 속성.|

```python
# 코모란과 okt 클래스를 사용해보도록하자.
txt = '''[강병수 / 낚시꾼 구조 시민 : 거기 세 분 중에 다 내가 들어갈게. 내가 들어갈게. 이렇게 말씀하셨는데, 그래도 거기서 제가 나이가 제일 어려서, 제가 들어가겠다고 하고.]
무사히 구조를 마친 강 씨, 그런데 다시 바다로 뛰어듭니다.
뒤집힌 보트까지 꺼내옵니다.
구조를 지켜보던 일행이 화면으로 보기에도 이상하리 만치 침착했던 건 모두 수상안전 전문가였기 때문이었습니다.
구조된 남성은 일행이 가져온 옷으로 체온을 유지해 안정을 취했고, 무사히 귀가한 것으로 전해졌습니다.'''
```

```python
# 1) okt 사용해보기 
from konlpy.tag import Okt
okt = Okt()

# 형태소 단위로 토큰화 해보자 
tokens = okt.morphs(txt)
print('- 토큰 갯수 : ',len(tokens))
print(tokens)

# 명사만 추출해보자 
noun_tokens = okt.nouns(txt)
print('- 명사 갯수 : ',len(noun_tokens))
print(noun_tokens)

# 품사를 확인해보자 
pos_tokens = okt.pos(txt)
print('- 품사확인 : \n',pos_tokens)

# 단어와 품사를 같이 봐보자 
pos_tokens2 = okt.pos(txt, join=True)
print('- 품사 확인2 :  \n', pos_tokens2)

# 해당 품사가 어떤 것지 확인해보자 
okt.tagset

# norm은 Okt만 있는 기능 : norm=True 
print(okt.morphs('이것도 되나욬ㅋㅋㅋㅋㅋ'))
print(okt.morphs('이것도 되나욬ㅋㅋㅋㅋㅋ', norm=True))
print(okt.morphs('반갑습니당',norm=True))
```

```python
#2) Komoran 
from konlpy.tag import Komoran
kom = Komoran()

tokens = kom.morphs(txt)
noun_token = kom.nouns(txt)
pos_token = kom.pos(txt)
kom.tagset
# kom.morphs(txt, norm = True) :  norm은 Okt에만 있다. 

print('- 토큰 갯수 : ',len(tokens))
print('- 명사 갯수 : ',len(noun_token))
print('- 단어 10개 품사확인 : \n',pos_token[:10])
print('- 품사 확인 : \n',kom.tagset)
```

## 62.4 연습1 - 코엔엘파이말뭉치 --> kolaw, kobill 로 예시
- 코엔엘파이 말뭉치로 kolaw, kobill이 있다. 
    - kolaw : 대한민국 헌법 말뭉치 
        - constitution.txt
    - kobill : 대한민국 국회 의안 말뭉치 
        - 1809890.txt ~ 1809899.txt
    
    

```python
from konlpy.corpus import kolaw, kobill

# 1) kolaw 확인

kolaw.fileids() # kolaw에 존재하는 파일 - ['constitution.txt'] 
constitution = kolaw.open('constitution.txt') # 파일열어서 로드
constitution.read() # 읽어보기 

# 2) kobill 확인
kobill.fileids() # kobill에 존재하는 파일들 
bill = kobill.open('1809896.txt')
bill.read()
```

- 프로세스 : 
    - (1) 임포트 및 객체생성
    - (2) 토큰화 
    - (3) 처리 
    - (4) 시각화 : matplotlib, dispersion_plot, WordCloud 

```python
from nltk import Text
from konlpy.corpus import kolaw
from konlpy.tag import Komoran

kom = Komoran()
kolaw.fileids()
law = kolaw.open('constitution.txt').read()
law_nouns = kom.nouns(law) # 명사추출 
print('- ', len(law_nouns))

text = Text(law_nouns, name = '헌법')
print(f'대한민국 : {text.count("대한민국")}, 헌법 : {text.count("헌법")}, 국민 : {text.count("국민")}')
```

```python
# 시각화 1 - matplotlib 
import matplotlib.pyplot as plt 
import matplotlib as mpl # 한글처리위해 

mpl.rcParams['font.family'] = 'AppleGothic'
plt.figure(figsize=(5,5))

text.plot(20) # 상위20개만 
plt.show()
```

```python
# 시각화 2 - dispersion_plot 
plt.figure(figsize=(5,5))
text.dispersion_plot(['법률', '대통령','국가','국회','헌법'])
```

```python
# 시각화 3 - word Cloud 
# !pip install wordcloud
from wordcloud import WordCloud

freq = text.vocab() # 빈도수 
print(freq)
print(f'총 단어 : {freq.N()}, 고유단어 : {freq.B()}, 최빈단어 : {freq.get(freq.max())}, 최빈단어 출현비율 : {freq.freq(freq.max())}')

font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf'
wc = WordCloud(font_path = font_path,max_words = 200)
word_cloud = wc.generate_from_frequencies(freq)
# word_cloud.to_file('word_cloud_law.png')

plt.figure(figsize=(10,10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```

```python
# 매개변수로 전달한 단어가 나오는 라인을 출력 
# 국민이나오는 50글자 만큼 라인을 10줄을 출력한다. 
text.concordance('국민', width = 50, lines = 10)
```
