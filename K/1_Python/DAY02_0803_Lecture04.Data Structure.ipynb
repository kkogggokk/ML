##  List Operator 
- 리스트 합치기 : + 
- 리스트 내 해당 인덱스 값 바꾸기 : 리스트명[인덱스] = 값 
- 곱하기로 여러개를 여러번 합치기 
- 리스트 안에 요소가 있는지 확인 : 요소 in 리스트명 
- 길이 구하기 : len(리스트명) <-- 내장함수형태 
- 맨 뒤에 요소 추가 하기 : .append(값) <-- 메소드 형태 
- 맨 뒤에 리스트 추가 : .extend(값) <-- 메소드형태 
- 원하는 곳에 삽입 : .insert(인덱스, 값) <-- 메소드 형태 
- 요소 삭제 : del 리스트명[인덱스] <-- 예약어형태 
- 원하는 값을 삭제 : .remove('값') <-- 메소드 형태 

```python
a = [1, 2, 3, 4]
b = [5, 6, 7, 8]
```

```python
# 리스트 합치기 
a + b 
```

```python
# 리스트 내 해당인덱스 값 바꾸기 
a[0] = 'something'
a
```

```python
# 곱하기로 여러개를 여러번 합치기 
a * 2 
```

```python
# 리스트 안에 요소가 있는지 확인 
'something' is a 
```

```python
seq = [1, 2, None, True]
```

```python
# 리스트 길이 구하기 , 내장함수 형태 , seq.length 가 아니다 
len(seq)
```

```python
# 맨 뒤에 요소 추가 , 메소드 형태 
seq.append('sth')
seq
```

```python
# 맨 뒤에 리스트 추가
seq.extend([5,6])
seq
```

```python
# 원하는 곳에 삽입 , 1번째 인덱스에 1.5 삽입
seq.insert(1, 1.5)
seq
```

```python
# 1번째 요소 삭제, 예약어 형태, seq.delete가 아니다
del seq[1]
seq
```

```python
#원하는 값을 삭제 
seq.remove('sth')
seq
```

## List as Mutable Data Type
- 리스트 등은 가변 타입이므로 주의해야한다. 

```python
'''
리스트는 가변타입이다. 
b = a 같은 주소를 가르키게 된다. 
a += [6] 인플레스 연산으로 기존객체에 할당하게 된다. 
b를 출력하면 같은 주소를 가르키기 때문에 [6]도 포함해서 출력된다. 
'''
a = [1, 2, 3, 4, 5]
b = a 
a += [6]
b 
```

```python
'''
str은 불변타입이다. 
b = a 로 같은 주소를 가르키게 된다. 
불변타입에 새로 대입하게 되므로 아웃플레이스 a = a+'?'로 새로운 객체를 만든다
a와 b는 출력값이 다르게 된다. 
'''
a = 'sth'
b = a 
a += '?'
print(a)
print(b)
```

## Packing & UnPacking
- 패팅 : 여러 데이터를 묶기 
    - t = 1, 2, 3, 4, 5
- 언패팅 : 묶인 데이터를 풀기
    - a, b, c, d, e = t 
    - \* : asterisk , 남는 요소를 리스트로 남기기 가능 

```python
t = [1, 2, 3, 4, 5]
a, b, c, _, _ = t # unpacking , _는 관습적으로 사용하지 않는 변수에 사용 
c 
```

```python
_
```

```python
a, *b, c = t # *로 남는 부분을 리스트로 묶을 수 있음 
```

```python
a, b, c
```

## Dictonary 
- 일종의 매핑을 위한 데이터 구조 
- key : value 형태로 구현 
- 불변타입으로만 이루어져 있으므로 key로 사용가능 
- {key1:va1, key2, val2, ...} 형태로 선언 

```python
dict = {
    1 : 'sth',
    (1, 2.5) : 1.5, # 튜플 --> 불변타입 --> hashale 할 경우 키로 사용 가능
    'text':2,
}
```

```python
dict[1]
```

```python
dict['text']
```

```python
dict[1, 2.5] # 튜플이므로 괄호 생략가능 
```

# set 
- 딕셔너리의 키만 모여 있는 형태 --> 집합형 
- 빈 셋 선언 : set(), 
- {} 은 빈 딕셔너리
- 요소 추가 , 중복된 요소는 추가 하지 않는다. : .add(값)
- 요소 삭제 , 존재하지 않는 요소 삭제 시 에러 발생 : .remove(값)
- 요소 삭제 , 존재하지 않을 경우 무시 : .discard(값)
- 여러 요소 추가, 중복은 무시 : .update([값1, 값2, ...])
- set 비우기 : .clear()

```python
s = set([1, 2, 3, 'text'])
s
```

```python
s.add(4)
s.add('text')
s
```

```python
s.remove(2)
```

```python
s.remove(99)
```

```python
s.discard(99)
s
```

```python
s.update([1, 99, None, True])
s
```

```python
s.clear()
s
```

## set Operations
- 수학적 집합 연산자가 존재 
- 교집합 : &
- 합집합 : |
- 차집합 : - 
- 배타적 합집합 : ^ 


```python
s1 = set([1, 2, 3, 4])
s2 = set([3, 4, 5, 6])
```

```python
s1 & s2 
```

```python
s1 | s2 
```

```python
s1 - s2 
```

```python
s2 - s1
```

```python
s1 ^ s2 
```
