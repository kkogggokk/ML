## conditional Statements
- 특정 조건이 만족될 경우 실행할 문항을 설정한다
- 들여쓰기와 : 으로 구문을 구분한다
- 들여쓰기의 convention은 스페이스 4칸 , 탭을 눌러 삽입 


- if[조건] : 조건을 검사하여 블록을 실행
- elif[조건] : 이전 조건과 맞지 않을 경우 조건을 다시 검사 및 실행
- else: 이전 모든 조건이 맞지 않을 경우 실행 

```python
if False:
    print('this sentence does not show')
if "":
    print('empty is false')
if 0 :
    print('0 is false')
if None:
    print('none is also false')
```

- 비교 연산자와 논리 연산자를 사용 

```python
a, b = 5, 8 
if a == 5 and b == 6:
    print('this is false')
if not a < b < 9 :
    print('this is false')
if a + 3 == b : 
    print('this is true')
```

## Ternary operators(삼항 연산자)
- 값1 if 조건 else 값2
- 조건이 참이면 값1을, 거짓이면 값2을 반환
- 연산자이다. 
- 블록으로 구분되는 문법요소가 아님

```python
value = 32 
'odd' if value % 2 else 'even' # 32를 2로 나눴을때 나머지가 --> 1(참)이면 'odd(홀수), 0(거짓)이면 'even(짝수)'
```

```python
('odd' if value % 2 else 'even') + '_number'
```

## while Statement
- 조건을 만족할때까지 출력 

```python
i = 1 # 초기값
while i < 10 : # 10보다 작을때 까지 반복 
    print(i) 
    i += 1
```

## for Statement
- python의 for문은 주어진 객체를 순환하는 개념
- for 요소 in [iterable]: 형태로 사용

```python
for i in [0, 1, 2, 3, 4]:
    print(i)
```

```python
for i in [0, 1, 2, 3, 4]:
    if i % 2 : 
        print(i, 'is odd')
```

## notions for loop
- 반복문의 변수명은 일반적으로 i, j, k로 지정 
- 반복문을 포함해서 프로그래밍에서 숫자의 시작은 대부분 0부터
- 반복문이 끝나지 않는 무한 루프에 주의
- 모든 순환이 가능한 객체는 for문을 적용하는 것이 가능 

```python
for c in 'this is text':
    print(c)
```

```python
for word in ['korea','hello','world']:
    print(word)
```

```python
for key in {'text':1, 'word':2}:
    print(key)
```

```python
for key,value in {'text':1, 'word':2}:
    print(value)
```

## controlling loop - break , continue 
- break문으로 가장 바깥의반복문을 나가는 것이 가능
- continue문으로 가장 바깥의 반복문의 처음으로 되돌아 가기 가능 

```python
# break문으로 가장 바깥의반복문을 나가는 것이 가능
for i in range(100):
    if i % 17 == 0 :
        break # i가 17보다 커지면 반복 나가기 
    print(i)
```

```python
# continue문으로 가장 바깥의 반복문의 처음으로 되돌아 가기 가능
for i in range(100):
    if i % 17:
        continue # i가 17의 약수가 아니면 for처음으로 
    print(i) # 17의 배수 출력 
```

## controll loop - else
- else 문으로 반복을 완전히 돌았을 경우 실행되는 블록 지정 가능 
- break로 중간에 나오게 되면 else문이 실행되지 않음 

```python
# else 문으로 반복을 완전히 돌았을 경우 실행되는 블록 지정 가능
for i in range(10):
    print(i)
else:
    print('loop complete with break')
```

```python
# break로 중간에 나오게 되면 else문이 실행되지 않음
for i in range(10):
    print(i)
    if i > 5 :
        break
else:
    print('loop complete without break')
```
