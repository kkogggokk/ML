## Assigning Variables 
- C언어와 달리 대입 연산이 딱히 반환 값을 가지는 것은 아니다.

```python
#C언어와 달리 대입 연산이 딱히 반환 값을 가지는 것은 아니다.
(a = 2) == 2
```

```python
# 연속해서 대입 가능 
a = b = 2
```

## Immutable Types & Mutable Types
- 파이썬에서 대입은 원칙적으로 메모리 주소복사 (즉, 값을 복사하지 않고 같은 주소를 공유)) 
- 불변형의 경우 수정이 필요한 경우에는 새로운 객체를 생성한다. 

```python
'''
int는 불변타입이다. 
a와 b는 같은 메모리 주소를 가르킨다. 
a는 불변타입이기 때문에 a=a+1를 실행하기위해 객체를 새로 생성해서 할당한다. 
a가 새로 할당되었기 때문에 a 와 b는 같지 않다'''
a = 10 
b = a 
a += 1 
a, b, a is b 
```

```python
'''
list는 가변타입니다.
a와 b는 같은 메모리 주소를 가르킨다. 
a는 가변타입으로 원 객체를 수정하게 된다. 
a의 메모리 주소는 변함없으므로 a 와 b 는 같다 
'''
a = [1, 2, 3]
b = a 
a += [4]
a, b, a is b 
```

```python
'''
a에 [5]를 추가한 객체를 생성해서 a에 할당한다.
a가 새로 할당되었기 때문에 a와 b는 다르다'''
a = a + [5]
a, b, a is b
```

# Features of Data Assignment 
- 파이썬에서 적당한 크기의 Primitive Data대입은 기존객체를 할당한다. 
- 불변타입이라 상관없음 

```python
'''
int는 Primitive Type이다 
b에 새로할당한 것처럼 보이지만 기존 객체를 가져온다
a와 b는 같은 메모리 주소를 가지는지 확인한다. '''
a = 1 
b = 1 
a is b 
```

```python
'''
int는 Primitive Type이지만 복잡한 값을 가짐
b에 값을 새로 할당한다. 
a와 b는 같은 메모리 주소를 가지는지 확인한다.'''
a = 1234567
b = 1234567
a is b 
```
