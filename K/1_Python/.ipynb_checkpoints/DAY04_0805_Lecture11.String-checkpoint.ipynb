# Recap Python String
- 파이썬 스트링의 특징
- 원시 자료형, 불변타입 
- 큰따옴표(") 혹은 작은따옴표(')로 표기된다.
- 따옴표를 3개 연달아 쓰면 여러줄을 넣을 수 있다. 
- 인덱싱 및 슬라이싱이 가능하다.
- 덧셈 및 곱셈이 가능하다.
- in & not in 연산이 가능하다.
    - 튜플과 다르게 작동
- 유니코드로 처리된다. 

```python
text = '''
여러줄 적기 가능 
docstring 
'''

print(text)
```

# Special Characters 
- 이스케이프문자(\)를 사용하여 특수문자 활용 


```python
text = '\
이렇게 적으면 \
엔터 없이 \
여러 줄을 적을수 있어요'
text
```

```python
text = '프린트할 때랑 \n평가할때랑 달라요'
print(text)
```

# Raw String
- r"<텍스트>"형태
- \을 무시하고 문자 그대로 취급 가능
- 주로 정규표현식에서 사용 

```python
string = '여기서는 역슬래시는 \n특별한 의미를 가져요'
string
```

```python
print(string)
```

```python
raw_str = r'여기서는 역슬래는 \n 특별한 의미를 가지지 않아요.'
raw_str
```

```python
print(raw_str)
```

```python
locate = 'C:\\Users\\user1'
print(locate)
```

```python
locate = r'C:\Users\user1'
print(locate)
```

# String Functions


```python
text = 'this is the python course'

print('[-]len        : ',len(text))
print('[-]upper      : ',text.upper())
print('[-]lower      : ',text.lower())
print('[-]capitalize : ',text.capitalize())
print('[-]title      : ',text.title())
```

```python
text = '      공백이 \t  있어요  \t\n     '

print('[-]strip  : ',text.strip())
print('[-]lstrip : ',text.lstrip())
print('[-]rstrip : ',text.rstrip())
```

```python
print('12345'.isdigit())
print('1.24e-3'.isdigit())
print('Capitalize'.isdigit())
print('lower_case'.islower())
```

# String Pattern Matching


```python
text = 'abc_text_abc_ee'
text1 = 'test_abc_text_abc_ee_abc'
pattern = 'abc'

print(text.count(pattern))
print(text.find(pattern))
print(text.rfind(pattern))
print(text.startswith(pattern))
print(text.endswith(pattern))

print(text1.find(pattern))
print(text1.startswith(pattern))
print(text1.endswith(pattern))
```
